```python
x1 = np.arange(0,11)
y1 = x**3

x2 = np.arange(0,11)
y2 = (x**3)/2

plt.plot(x1,y1, linewidth=5) #linewidth sets the width of the line in the line chart
plt.plot(x2,y2, linewidth=5)
plt.title('Multiple Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

- Adding/Editing Colors/Legends to the Chart

```python
x1 = np.arange(0,11)
y1 = x**3

x2 = np.arange(0,11)
y2 = (x**3)/2

plt.plot(x1,y1, 'r', label = 'x1,y1 data', linewidth=5) 
plt.plot(x2,y2, 'b', label = 'x2,y2 data', linewidth=5)
plt.legend() #needed for adding label
plt.title('Multiple Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.figure(figsize=(15,5))
years=[1950,1955,1960,1970,1975,1980]
pops = np.array([1262,1650,2525,2758,3018,3322])
pops_b = pops/1000
plt.plot(years, pops_b, linewidth = 3)
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
plt.figure(figsize=(15,5))
years=[1950,1955,1960,1970,1975,1980]
pops = np.array([1262,1650,2525,2758,3018,3322])
pops_b = pops/1000
deaths = [1.0,2.0,2.4,2.6,3.9,5.0]
plt.plot(years, pops_b, linewidth = 3)
plt.plot(years, deaths, 'r', label = "Mortality Rate", linewidth = 5)
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
plt.figure(figsize=(15,5))
years=[1950,1955,1960,1970,1975,1980]
pops = np.array([1262,1650,2525,2758,3018,3322])
pops_b = pops/1000
deaths = [1.0,2.0,2.4,2.6,3.9,5.0]
plt.plot(years, pops_b, linewidth = 3)
plt.plot(years, deaths, 'r', label = "Mortality Rate", linewidth = 5)
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.legend()
plt.grid(True) #for grids
```

#### Inserting marker and markersize

```python
plt.figure(figsize=(15,5))
years=[1950,1955,1960,1970,1975,1980]
pops = np.array([1262,1650,2525,2758,3018,3322])
pops_b = pops/1000
deaths = [1.0,2.0,2.4,2.6,3.9,5.0]
plt.plot(years, pops_b, 'g', label = 'Population Trends', marker='*', markersize=10,linewidth = 3)
plt.plot(years, deaths, 'y', label = "Mortality Rate", marker='o', markersize=10, linewidth = 5)
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.legend()
plt.grid(True)
```

### Pie Charts

```python
raw_data = {'names':['Nick', 'Sani', 'John', 'Rubi', 'Maya'],
           'jan_score':[123,124,125,126,128],
           'feb_score':[23,24,25,27,29],
           'march_score':[3,5,7,6,9]}

df = pd.DataFrame(raw_data, columns=['names', 'jan_score', 'feb_score', 'march_score'])
df['total_score']=df['jan_score']+df['feb_score']+df['march_score']
df
```

```python
plt.pie(df['jan_score'], labels=df['names'], autopct='%.2f%%')
#autopct='%.2f%%' gives percentage upto two places
plt.axis('equal')
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['jan_score'], labels=df['names'], autopct='%.2f%%', explode=(0.5,0,0,0,0))
plt.axis('equal')
plt.axis('equal')
plt.show()
#explode shows a seperation
```

```python
plt.pie(df['jan_score'], labels=df['names'], autopct='%.2f%%', explode=(0.5,0.5,0,0,0))
plt.axis('equal')
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['jan_score'], labels=df['names'], autopct='%.2f%%', explode=(0.5,0.5,0.5,0,0))
plt.axis('equal')
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['jan_score'], labels=df['names'], autopct='%.2f%%', explode=(0.5,0.1,0,0,0))
plt.axis('equal')
plt.axis('equal')
plt.show()
#numeric value in explode shows the distance of seperation
```

### SUBPLOTS

- Horizontal Subplots

```python
plt.figure(figsize=(15,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

plt.subplot(1,2,1) #in this, first two parameters i.e. 1,2 denotes we are creating a subplot of 1 row 2 columns and the last parameter 1 indicates 1st graph
plt.plot(x,y,'g')
plt.title("Fisrt Graph")
 
plt.subplot(1,2,2) #1 row, 2 columns and the 2nd graph
plt.plot(y,x,'b')
plt.title("Second Graph")
```

```python
plt.figure(figsize=(15,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

plt.subplot(1,3,1) 
plt.plot(x,y,'g')
plt.title("Fisrt Graph")
 
plt.subplot(1,3,2) 
plt.plot(y,x,'b') #changing x&y position leads to horizontal
plt.title("Second Graph")

plt.subplot(1,3,3) 
plt.plot(x**2,y,'y')
plt.title("Third Graph")
```

```python
plt.figure(figsize=(15,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

plt.subplot(1,3,1) 
plt.plot(x,y,'g')
plt.title("Fisrt Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
 
plt.subplot(1,3,2) 
plt.plot(y,x,'b') #changing x&y position leads to horizontal
plt.title("Second Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.subplot(1,3,3) 
plt.plot(x**2,y,'y')
plt.title("Third Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.tight_layout(w_pad=10) #this allows spoce between graphs and w_pad stands for width padding
```

### Vertical Subplots

```python
plt.figure(figsize=(15,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

plt.subplot(2,1,1) #2rows, 1 column
plt.plot(x,y,'g')
plt.title("Fisrt Graph")
 
plt.subplot(2,1,2) #2rows, 1 column and the 2nd graph
plt.plot(y,x,'b')
plt.title("Second Graph")
```

```python
plt.figure(figsize=(15,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

plt.subplot(2,2,1) 
plt.plot(x,y,'g')
plt.title("Fisrt Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
 
plt.subplot(2,2,2) 
plt.plot(y,x,'b') #changing x&y position leads to horizontal
plt.title("Second Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.subplot(2,2,3) 
plt.plot(y*2,x*2,'y')
plt.title("Third Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.subplot(2,2,4) 
plt.plot(x*2,y*2,'r')
plt.title("Fourth Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.tight_layout(h_pad = 5, w_pad=10)
```
